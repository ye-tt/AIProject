# 1. 导数，偏导数
## 1.1 导数
**定义**

导数描述的是一个单变量函数 f(x) 在某一点 x 处的瞬时变化率，或者说函数曲线在该点的切线斜率。

* 记法：$f'(x)$ 或 $df/dx$ 或 $∂f/∂x$ (在单变量时，偏导符号也常用)。
* 数学定义：
> $f'(x) = lim_{(h->0)}[f(x + h) - f(x)] / h$
这表示当自变量 x 的改变量 h 趋近于 0 时，函数值的改变量与 x 的改变量之比的极限。

**直观理解**

* 斜率：导数就是函数图像在某点切线的斜率。
* 速度：如果 f(t) 表示物体在时间 t 的位置，那么 f'(t) 就是物体在时间 t 的瞬时速度。
* 增长/减少：
f'(x) > 0：函数在 x 点附近是递增的。
f'(x) < 0：函数在 x 点附近是递减的。
f'(x) = 0：函数在 x 点可能是极值点（极大值或极小值）或拐点。


## 1.2 偏导数

偏导数（Partial Derivative） 是多元函数微分学中的核心概念，用于描述当一个多变量函数的其他变量保持不变时，函数关于其中一个变量的变化率

直观理解
想象你在一座山（地形由函数 $z=f(x,y)$ 描述）上
偏导数 $∂f/∂x$：表示你只沿东西方向（x 方向）走，海拔变化的快慢（坡度），而南北位置（y）不动。
偏导数 $∂f/∂y$：表示你只沿南北方向（y 方向）走，海拔变化率，东西位置（x）固定

**定义**

设函数$f(x_1,x_2,...,x_n)$是一个 n 元函数。

函数 $f$ 关于第 i 个变量 $x_i$ 的偏导数定义为：

$\frac{​∂f}{​∂x_i​ }= lim_{h->0} \frac{f(x_1​,…,x_i+h,...x_n)​−f(x_1​,…,x_i,...x_n)​}{h}$
